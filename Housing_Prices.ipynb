## Cleaning

```python
train = train.drop(['Id', 'MiscFeature'], axis=1)
test = test.drop(['MiscFeature'], axis=1)
```

```python
mean = ['GarageYrBlt']
zero = ['LotFrontage', 'MasVnrArea']
none = ['Alley', 'MasVnrType', 'BsmtQual', 'BsmtCond', 'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2', 'Electrical', 'FireplaceQu', 'GarageType', 'GarageFinish', 'GarageQual', 'GarageCond', 'PoolQC', 'Fence']

for col in zero:
    train[col + '_input'] = train[col].isna().astype(int)
    train[col].fillna(0, inplace=True)

for col in none:
    train[col + '_input'] = train[col].isna().astype(int)
    train[col].fillna('None', inplace=True)

for col in mean:
    train[col + '_input'] = train[col].isna().astype(int)
    train[col].fillna(train[col].mean(), inplace=True)
```

```python
most_common = ['MSZoning', 'Utilities', 'Exterior1st', 'Exterior2nd', 'KitchenQual', 'Functional', 'MSZoning', 'SaleType']
zero = ['LotFrontage', 'MasVnrArea', 'BsmtFullBath', 'TotalBsmtSF', 'BsmtHalfBath']
none = ['Alley', 'MasVnrType', 'BsmtQual', 'BsmtCond', 'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2', 'Electrical', 'FireplaceQu', 'GarageType', 'GarageFinish', 'GarageQual', 'GarageCond', 'PoolQC', 'Fence']
mean = ['BsmtFinSF1', 'BsmtFinSF2', 'BsmtUnfSF', 'GarageYrBlt', 'GarageArea', 'GarageCars']

for col in most_common:
    test[col + '_input'] = test[col].isna().astype(int)
    most_common = test[col].mode()[0]
    test[col].fillna(most_common, inplace=True)

for col in zero:
    test[col + '_input'] = test[col].isna().astype(int)
    test[col].fillna(0, inplace=True)

for col in none:
    test[col + '_input'] = test[col].isna().astype(int)
    test[col].fillna('None', inplace=True)

for col in mean:
    test[col + '_input'] = test[col].isna().astype(int)
    test[col].fillna(train[col].mean(), inplace=True)
```

```python
dummies = []

for col in train.select_dtypes(include='object').columns:
    if train[col].nunique() > 15:
        train[col], mapping = pd.factorize(train[col])
        mapping_dict = dict(enumerate(list(mapping)))
        test[col] = test[col].map(lambda x: next(key for key, value in mapping_dict.items() if value == x))
    else:
        dummies.append(col)

train = pd.get_dummies(train, columns=dummies)
test = pd.get_dummies(test, columns=dummies)
```

```python
x_train = train.drop("SalePrice", axis=1)
y_train = train["SalePrice"]
x_test = test.drop("Id", axis=1)
HouseId = test["Id"]
```

```python
missing_columns = set(x_train.columns) - set(x_test.columns)

for col in missing_columns:
    x_test[col] = 0
```

```python
missing_columns = set(x_test.columns) - set(x_train.columns)

for col in missing_columns:
    x_train[col] = 0
```

# Training Model

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error
```

## Random Florest

```python
def testing():
    everything = []

    options = []

    for n in range(100, 301, 10):
        for depth in range(8, 25, 1):
            for mss in range(2, 4):
                options.append([n, depth, mss])

    for op in options:
        sum = 0

        for num in range(0, 5):
            train_x, val_x, train_y, val_y = train_test_split(x_train, y_train, random_state = num, test_size = 0.2)
            model = RandomForestRegressor(random_state = 0, n_estimators = op[0], max_depth = op[1], min_samples_split = op[2])
            model.fit(train_x, train_y)
            preds = model.predict(val_x)
            sum += round(mean_absolute_error(val_y, preds), 2)

        everything.append({f'{op}': round(sum/5, 2)})

    return everything
```

```python
ar = testing()
print(ar)
```

```python
max = {'irrelevant': 30000}

for obj in ar:
    if list(obj.values())[0] < list(max.values())[0]:
        max = obj

print(max)
```
